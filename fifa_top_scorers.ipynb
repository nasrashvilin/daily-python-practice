```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
df = pd.read_csv('data/wc_top_scorers.csv')
```

```python
# inspect
df.head()
df.info()
df.describe()
```

```python
#plot most goals scored by players in the FIFA World Cup

sns.set_style("dark")
plt.figure(figsize=(10, 6))


ax = sns.barplot(data=df, x='goals',
            y='player',
            order=df.sort_values('goals', ascending=False)['player'].head(10),
            )
# Add labels to the bars
ax.bar_label(ax.containers[0], fontsize=12);

#remove spines
sns.despine(left=True, bottom=True)
plt.title('Top 10 FIFA World Cup Scorers')
plt.xlabel('Goals') 
plt.ylabel('Player')
plt.show()
```

```python
# Plot cantries with most top scorers at WC 

# Create dataframe with count of players by country
top_countries = df.groupby('country')['player'].count().sort_values().reset_index()

# create bar plot with matplotlib
# Set size of the plot
plt.figure(figsize=(10, 8))

plt.barh(top_countries['country'], top_countries['player'])
plt.title('Countries with Most Top Scorers at FIFA World Cup')
plt.xlabel('Number of Top Scorers')

plt.tight_layout()
plt.show()
```
