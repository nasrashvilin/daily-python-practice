```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

# For regression 
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
penguins = sns.load_dataset('penguins')
penguins.head(5)
```

```python
penguins = penguins[['body_mass_g', 'bill_length_mm',
                     'sex', 'species']]
penguins.dropna(inplace = True)
```

```python
penguins.describe()
penguins.head(5)
```

```python
# select the independent variables or features
penguins_x = penguins[['sex', 'species', 'bill_length_mm']]

# select the dependent variable
penguins_y = penguins['body_mass_g']
```

```python
from sklearn.model_selection import train_test_split

# devide data into training and holdout datasets
X_train, X_test, y_train, y_test = train_test_split(penguins_x, penguins_y,
                                                    test_size = 0.3, random_state = 42)

```

```python
#regression formula 
ols_formula = 'body_mass_g ~ bill_length_mm + C(sex) + C(species)' # C() for categorical variables

ols_data = pd.concat([X_train, y_train], axis = 1)
```

```python
OLS = ols(ols_formula, data = ols_data)

model = OLS.fit()

model.summary()
```
